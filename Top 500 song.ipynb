# Import librairies:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Dataset:

```python
df= pd.read_csv(r"C:\Users\user\Desktop\projet python\songs.csv",sep=",", encoding='Latin-1')
df =df.dropna()
df
```

# Analysis:

## Modify columns Content:

### Position and Year:

```python
df['position']=df['position'].str[3:]
df['Year']=df['released'].str[-4:]
df['streaklg']=df['streak'].str[:-6]
```

## EDA:

### Top Songs:

```python
df_top=df.groupby('Year').count().reset_index()
df_top
plt.figure(figsize=(20,10))
plt.bar(df_top['Year'],df_top['title'])
plt.xticks(rotation = 'vertical')
plt.xlabel('Years')
plt.ylabel('Recurrences')
plt.title('Numbers of Top Song By Year')
```

### Top Artists:

```python
df_cnt =df.groupby('artist').count().sort_values(by='description',ascending = False).reset_index()
df_cnt = df_cnt[:10]

plt.figure(figsize=(15,8))
plt.pie(df_cnt['title'],labels = df_cnt['artist'],autopct='%.2f',shadow=True)
plt.title("Top Contributers")

j=0
for i in df_cnt['artist']:
    print("The",j+1,"Aritst is :",df_cnt['artist'][j])
    j=j+1
```

### Top Writers:

```python
df_w=df.groupby('writers').count().sort_values(by='description',ascending = False).reset_index()
df_w = df_w[:10]
plt.figure(figsize=(15,8))
plt.pie(df_w['title'],labels = df_w['writers'], autopct = "%.2f",shadow = True)
plt.title("Top Writers")

j=0
for i in df_w['writers']:
    print("The",j+1,"Writer is :",df_w['writers'][j])
    j=j+1
```

### Top Producers:

```python
df_p = df.groupby('producer').count().sort_values(by='description',ascending = False).reset_index()
df_p = df_p[:10]
plt.figure(figsize=(15,8))
plt.pie(df_p['title'],labels = df_p['producer'],autopct = '%.2f',shadow = True)
plt.title("Top Producers")
j=0
for i in df_p['producer']:
    print("The",j+1,"Producer is :",df_p['producer'][j])
    j=j+1
```

### Top Streaks:

```python

#conversion to int 
df['streaklg']=pd.to_numeric(df['streaklg'],errors='coerce')
df_streak=df.sort_values(by='streaklg',ascending = False)
df_streak=df_streak.dropna()
df_streak['streaklg']=df_streak['streaklg'].astype(int)

j=0
for i in range(0,10):
    print(df_streak['title'][j],'** by **',df_streak['writers'][j],'**made a streak of**',df_streak['streak'][j],'.')
    j=j+1

    
plt.figure(figsize=(30,10))
plt.bar(df_streak['title'][:20],df_streak['streaklg'][:20])
plt.xticks(rotation='vertical')
plt.xlabel('Song')
plt.ylabel('Weeks')
```
